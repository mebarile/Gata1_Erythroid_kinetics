```python
import numpy as np
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
import pyreadr
import scvelo as scv
```

```python
# Read data abd prepare objects
```

```python
adata_gata = sc.read('./adata_gata1.h5')
adata_atlas = sc.read('./adata_umap_pca.h5')
```

```python
df_ann = pd.read_csv('df_annotation_all_bat.csv', index_col = 0)
```

```python
barcode_t = [str.split(cellname, '_')[1] for cellname in df_ann.index]

barcode = [str.split(cellname, '-')[0] for cellname in barcode_t]
typ =  [str.split(cellname, '_')[0] for cellname in df_ann.index]

lista = [x + ':' + y + 'x' for x,y in zip(typ,barcode)]

df_ann.index = lista

df_new = df_ann.reindex(adata_gata.obs_names)
```

```python
adata_gata.obs['stage_mapped'] = df_new['stage.mapped']
adata_gata.obs['celltype_mapped'] = df_new['celltype.mapped']
adata_gata.obs['closest_cell'] = df_new['closest.cell']
```

```python
adata_gata.obs['pop_haem_ref'] = list(adata_atlas.obs.haem_subclust[df_new['closest.cell']])
```

```python
corr_pca = pd.read_csv('./corrected_pca_gata_times.csv',index_col=0)

barcode_t = [str.split(cellname, '_')[1] for cellname in corr_pca.index]
barcode = [str.split(cellname, '-')[0] for cellname in barcode_t]
typ =  [str.split(cellname, '_')[0] for cellname in corr_pca.index]

lista = [x+':'+y+'x' for x,y in zip(typ,barcode)]

corr_pca.index = lista

pca = corr_pca.reindex(adata_gata.obs_names)
```

```python
# Compute UMAP from corrected PCAs
```

```python
adata_new = sc.AnnData(adata_gata.X)

adata_new.obs_names = adata_gata.obs_names
adata_new.obs = adata_gata.obs

adata_new.var = adata_gata.var

adata_new.obsm['X_pca'] = pca.values
```

```python
sc.pp.neighbors(adata_new,n_neighbors=20)

sc.tl.umap(adata_new)
```

```python
adata_rid = adata_new[(adata_new.obs['celltype_mapped'] == 'Endothelium') |
    (adata_new.obs['celltype_mapped'] == 'Haematoendothelial progenitors') |
       (adata_new.obs['celltype_mapped'] == 'Erythroid1') | 
       (adata_new.obs['celltype_mapped'] == 'Erythroid2') |
       (adata_new.obs['celltype_mapped'] == 'Erythroid3') | 
       (adata_new.obs['celltype_mapped'] == 'Blood progenitors 1') |
       (adata_new.obs['celltype_mapped'] == 'Blood progenitors 2'),:].copy()
```

```python
vec = (adata_rid.obs.pop_haem_ref == 'nan') | (adata_rid.obs.pop_haem_ref == 'Mes2')
adata_rid = adata_rid[~vec,:].copy()
```

```python
list1 = list(adata_rid.obs.tomato)
list2 = [('tdTom-' if entry == 'neg'else 'tdTom+' ) for entry in list1]

adata_rid.obs['tdTom'] = list2

new_col = adata_rid.obs.pop_haem_ref.astype('str') +'_'+ adata_rid.obs.tdTom.astype('str')

adata_rid.obs['comb'] = new_col
```

```python
list1 = list(adata_rid.obs.comb)
list_n = [(entry if entry.endswith('-') else 'others' ) for entry in list1]
list_n = [(entry if not entry.startswith('na') else 'others' ) for entry in list_n]
adata_rid.obs['names_neg'] = list_n
```

```python
list1 = list(adata_rid.obs.comb)
list_p = [(entry if entry.endswith('+') else 'others' ) for entry in list1]
list_p = [(entry if not entry.startswith('na') else 'others' ) for entry in list_p]
adata_rid.obs['names_pos'] = list_p

new_col = adata_rid.obs.stage_mapped.astype('str') +'_'+ adata_rid.obs.tdTom.astype('str')

adata_rid.obs['combt'] = new_col
```

```python
list_t = list(adata_rid.obs.combt)
list_nt = [(entry if entry.endswith('-') else 'others' ) for entry in list_t]
lista_nt = [(entry if not entry.startswith('EC') else 'others' ) for entry in list_nt]
lista_nt = [(entry if not entry.startswith('na') else 'others' ) for entry in list_nt]



adata_rid.obs['time_neg'] = list_nt

list_t = list(adata_rid.obs.combt)
list_pt = [(entry if entry.endswith('+') else 'others' ) for entry in list_t]
list_pt = [(entry if not entry.startswith('EC') else 'others' ) for entry in list_pt]
list_pt = [(entry if not entry.startswith('na') else 'others' ) for entry in list_pt]



adata_rid.obs['time_pos'] = list_pt
```

```python
green = "#caf75e"

color = ["#6460c5",
"#96b8e4",
"#07499f",
"#036ef9",
green,
green,
green,
green,
green,
green,
green,
green,
"#f67a58",
"#a26724",
"#cdaf7f",
"#625218",
"#bb22a7",
"#f695e9",
"#02f9ff",
"#4c4a81",
"#f6931d",
"#c62127",
"#E0E0E0"]

colorp = ["#6460c5",
"#96b8e4",
"#07499f",
"#036ef9",
green,
green,
green,
green,
green,
green,
green,
green,
"#f67a58",
"#a26724",
"#625218",
"#bb22a7",
"#f695e9",
"#02f9ff",
"#4c4a81",
"#f6931d",
"#c62127",
"#E0E0E0"]

color_t = ["#FDAE61",
"#FFFFBF",
"#FEE08B",
"#E6F598",
"#ABDDA4",
"#66C2A5",
"#3288BD"
]



ln = np.unique(list_n)

lnt = np.unique(list_nt)

lp = np.unique(list_p)

lpt = np.unique(list_pt)
```

```python
# Plot Figure 4E
```

```python
vec = [(True if pop in ['Haem1','Haem2', 'Haem3', 'Haem4','BP1', 'BP2', 'BP3', 'BP4','Ery1', 'Ery2', 'Ery3', 'Ery4','Mk', 'My'] else False) for pop in list(adata_rid.obs.pop_haem_ref)]

df_haemr = adata_rid.obs.loc[vec]
```

```python
df_n = df_haemr[df_haemr.index.str.startswith('SIGAB3')].groupby('stage_mapped').count()

df_a = df_haemr[df_haemr.index.str.startswith('SIGAA3')].groupby('stage_mapped').count()



df_poph = df_n.iloc[:,2:]
df_poph['pos'] = df_a['pop_haem_ref']

df_poph['neg'] = df_n['pop_haem_ref']

df_poph = df_poph.iloc[:,1:]

plt.bar(range(1,70,10),df_poph['pos']

,width=2,color='r')
plt.bar(range(2,72,10),df_poph['neg']
        
,width =2,color='k')
plt.xticks(rotation=90)
plt.xticks(range(1,70,10),('E7.0','E7.25','E7.5','E7.75','E8.0','E8.25','E8.5'));
plt.rcParams["axes.grid"] = False

plt.ylim([0, 900])
```

```python
df_n = adata_rid.obs[adata_rid.obs.index.str.startswith('SIGAB3')].groupby('pop_haem_ref').count()

df_a = adata_rid.obs[adata_rid.obs.index.str.startswith('SIGAA3')].groupby('pop_haem_ref').count()



df_poph = df_n.iloc[:,2:]
df_poph['pos'] = df_a['closest_cell']

df_poph['neg'] = df_n['closest_cell']


df_poph = df_poph.iloc[:,1:]

plt.bar(range(1,14*5,5),df_poph['pos'].loc[['Haem1','Haem2', 'Haem3', 'Haem4','BP1', 'BP2', 'BP3', 'BP4','Ery1', 'Ery2', 'Ery3', 'Ery4','Mk', 'My']]

,width=2,color='r')
plt.bar(range(2,14*5+2,5),df_poph['neg'].loc[['Haem1','Haem2', 'Haem3', 'Haem4','BP1', 'BP2', 'BP3', 'BP4','Ery1', 'Ery2', 'Ery3', 'Ery4','Mk', 'My']]

,width =2,color='k')
plt.xticks(rotation=90)
plt.xticks(range(1,14*5,5),('Haem1',
       'Haem2', 'Haem3', 'Haem4','BP1', 'BP2', 'BP3', 'BP4','Ery1', 'Ery2', 'Ery3', 'Ery4','Mk', 'My'));
plt.rcParams["axes.grid"] = False

plt.ylim([0, 900])
```
