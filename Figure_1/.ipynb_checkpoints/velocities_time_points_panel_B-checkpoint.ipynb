```python
import numpy as np
import scvelo as scv
import matplotlib.pyplot as plt
```

Prepare objects

```python
whole_celltype_colors = ["#532C8A",
'#c19f70',
"#f9decf",
"#c9a997",
"#B51D8D",
"#3F84AA",
"#9e6762",
"#354E23",
"#F397C0",
"#eda450",                             
"#635547",
"#C72228", 
"#f77b59",
"#EF4E22", 
"#8870ad",
"#647a4f",
"#EF5A9D",   
"#FBBE92",           
"#139992",  
"#ed8f84",
"#DFCDE4",
"#8EC792",                                    
"#C594BF",                 
"#C3C388",
"#0F4A9C",                          
"#FACB12",                                       
"#8DB5CE",
"#C9EBFB",  
"#DABE99",
"#65A83E",
"#005579",                                  
"#CDE088",
"#BBDCA8",
"#F6BFCB"]
```

```python
adata = scv.read('adata_umap_pca.h5')
```

```python
adata.uns['celltype_colors'] = whole_celltype_colors
populations_whole = np.unique(adata.obs.celltype)
```

```python
adata_e7_5 = adata[adata.obs.stage == 'E7.5']
adata_e8_5 = adata[adata.obs.stage == 'E8.5']
```

```python
populations_E7_5 = np.unique(adata_e7_5.obs.celltype)
populations_E8_5 = np.unique(adata_e8_5.obs.celltype)
```

```python
colors_E7_5 = [whole_celltype_colors[i] for i in range(len(whole_celltype_colors)) if populations_whole[i] in populations_E7_5]
colors_E8_5 = [whole_celltype_colors[i] for i in range(len(whole_celltype_colors)) if populations_whole[i] in populations_E8_5]
```

Compute velocities E7.5

```python
scv.pp.filter_and_normalize(adata_e7_5, n_top_genes = 5000, min_shared_counts = 20)
scv.pp.moments(adata_e7_5, n_pcs = 50, n_neighbors = 30)
```

```python
scv.tl.recover_dynamics(adata_e7_5) # This function was run as a separate script remotely because it is computational demanding 
```

```python
adata_e7_5 = scv.read('adata_e7_5.h5') # Object plus velocities
adata_e7_5.uns['celltype_colors'] = colors_E7_5
```

```python
scv.tl.velocity(adata_e7_5, mode = "dynamical")
scv.tl.velocity_graph(adata_e7_5)
```

```python
fig, axs = plt.subplots(1,1, figsize = (10,10))

scv.pl.velocity_embedding_stream(
    adata_e7_5, basis = "umap", color="celltype", size=50, ax = axs)

plt.show()
```

Compute velocities E8.5

```python
scv.pp.filter_and_normalize(adata_e8_5, n_top_genes = 5000, min_shared_counts = 20)
scv.pp.moments(adata_e8_5, n_pcs = 50, n_neighbors = 30)
```

```python
scv.tl.recover_dynamics(adata_e8_5) #  This function was run as a separate script remotely because it is computational demanding
```

```python
adata_e8_5 = scv.read('adata_e8_5.h5') # Object plus velocities
adata_e8_5.uns['celltype_colors'] = colors_E8_5
```

```python
scv.tl.velocity(adata_e8_5, mode = "dynamical")
scv.tl.velocity_graph(adata_e8_5)
```

```python
fig, axs = plt.subplots(1,1, figsize = (10,10))

scv.pl.velocity_embedding_stream(
    adata_e8_5, basis = "umap", color = "celltype", size = 50, ax = axs)

plt.show()
```
